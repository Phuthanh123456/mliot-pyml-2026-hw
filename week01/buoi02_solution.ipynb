```python
import math
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
sns.set_theme(style='whitegrid')

df = sns.load_dataset('iris')
df.head()
```

```python
print('Phan 1')

print('shape:', df.shape)
print(df.head())
print(df.dtypes)

num_cols = df.select_dtypes(include='number').columns
stats = df[num_cols].agg(['mean', 'median', 'var', 'std', 'min', 'max'])
stats.loc['mode'] = df[num_cols].mode().iloc[0]
q1 = df[num_cols].quantile(0.25)
q3 = df[num_cols].quantile(0.75)
iqr = q3 - q1
extra = pd.DataFrame({'q1': q1, 'q3': q3, 'iqr': iqr})
print(stats)
print(extra)
print(df.groupby('species')[num_cols].agg(['mean', 'std']))
```

```python
print('Phan 4')

p_b = 0.01
p_pos_given_b = 0.99
p_pos_given_nb = 0.05
p_pos = p_pos_given_b * p_b + p_pos_given_nb * (1 - p_b)
p_b_given_pos = p_pos_given_b * p_b / p_pos
print(round(p_b_given_pos, 4))

def naive_bayes_spam(message, spam_words, p_spam=0.5, p_word_spam=0.8, p_word_ham=0.2):
    words = message.lower().split()
    score_spam = math.log(p_spam)
    score_ham = math.log(1 - p_spam)
    for word in words:
        if word in spam_words:
            score_spam += math.log(p_word_spam)
            score_ham += math.log(p_word_ham)
        else:
            score_spam += math.log(1 - p_word_spam)
            score_ham += math.log(1 - p_word_ham)
    return 'spam' if score_spam > score_ham else 'ham'

spam_words = {'free', 'win', 'offer', 'prize', 'urgent'}
for msg in ['free prize now', 'meeting tomorrow', 'urgent offer just for you']:
    print(msg, '->', naive_bayes_spam(msg, spam_words))
```

```python
print('Phan 3')

cov = df[num_cols].cov()
corr = df[num_cols].corr()
print(cov)
print(corr)

plt.figure(figsize=(7, 5))
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0)
plt.tight_layout()
plt.show()

sns.pairplot(df, hue='species', corner=True, diag_kind='hist')
plt.show()
```
